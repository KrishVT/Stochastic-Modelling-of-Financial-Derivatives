```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
alpha = 0.1     
sigma = 0.2           
S0 = 100
```

```python
T = 1.0
N = 500
dt = T / N
```

```python
Current_Time = []

for curr in range(501):
    Current_Time.append(round(curr * dt, 6))

Current_Time
```

```python
def gbm():
    dw = []

    for i in range(501):
        dw.append(np.random.normal(0, np.sqrt(dt)))
        
    for i in range(501):
        if i != 0:
            dw[i] += dw[i - 1]
            
    for i in range(501):
        dw[i] = S0 * np.exp(sigma * dw[i] + (alpha - 0.5 * sigma**2) * Current_Time[i])
        
    return dw
```

```python
# Plotting 5 paths
plt.figure(figsize=(10, 6))

for _ in range(5):
    path = gbm()
    plt.plot(Current_Time, path)

plt.title("5 Paths of Geometric Brownian Motion")
plt.xlabel("Time $t$")
plt.ylabel("Stock Price $S(t)$")
plt.grid(True)
plt.show()
```
