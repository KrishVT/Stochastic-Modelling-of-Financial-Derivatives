```python
Necessary Libraries
import numpy as np
import matplotlib.pyplot as plt
```

```python
T = 1.0
Number_of_Points = 500
dt = T / Number_of_Points

Current_Time = []

for curr in range(501):
    Current_Time.append(round(curr * dt, 6))

Current_Time
```

```python
dw = []

for i in range(501):
    dw.append(np.random.normal(0, np.sqrt(dt)))
    
dw
```

```python
for i in range(501):
    if i != 0:
        dw[i] += dw[i - 1]
        
dw
```

```python
plt.figure(figsize=(10, 5))
plt.plot(Current_Time, dw, label="Wiener Process Path")
plt.title("Simulation of a Wiener Process (Brownian Motion)")
plt.xlabel("Time")
plt.ylabel("W(t)")
plt.grid(True)
plt.legend()
plt.show()
```

# The probability of the above attached Weiner Process Path is exactly zero as there are infinite number of paths or in simple words in a normal distribution the probability of a particular value is "0".
