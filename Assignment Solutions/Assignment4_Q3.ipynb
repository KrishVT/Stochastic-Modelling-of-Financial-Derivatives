***Estimation***

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
# Parameters
number = 10000           # Total drops
distance = 2.0           # Distance between lines
length = 1               # Needle length (l ≤ d)
```

```python
# Generate random x and θ
x = np.random.uniform(0, distance/2, number)
theta = np.random.uniform(0, np.pi/2, number)
```

```python
# Count number of crossings
crossings = np.sum(x <= (length / 2) * np.sin(theta))
```

```python
# Monte Carlo estimate of π
pi_estimate = (2 * number * length) / (distance * crossings)
print("Estimated π:", pi_estimate)
```

***Convergence***

```python
current_value_of_pi = []
cross_count = 0

for i in range(1, number+1):
    xi = np.random.uniform(0, distance/2)
    thetai = np.random.uniform(0, np.pi/2)
    if xi <= (length/2) * np.sin(thetai):
        cross_count += 1
    if cross_count > 0:
        current_value_of_pi.append((2 * i * length) / (distance * cross_count))
    else:
        current_value_of_pi.append(0)

plt.figure(figsize=(10, 5))
plt.plot(range(1, number+1), current_value_of_pi)
plt.axhline(np.pi, color='r', linestyle='--', label='True π')
plt.xscale('log')
plt.xlabel('Number of Drops (log scale)')
plt.ylabel('Estimated π')
plt.title("Convergence of π Estimate using Buffon's Needle")
plt.legend()
plt.grid(True)
plt.show()
```
